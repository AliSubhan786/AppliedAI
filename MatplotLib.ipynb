```python
import matplotlib.pyplot as plt
import numpy as np
#it treats indexes as x values if it isn't given
y = np.array([0,1,4,9,16,25,36,49,64,81,100])
plt.plot(y, 'r--')
plt.xlabel("x-Indices")
plt.ylabel("y-squares")
plt.title("My plot")
plt.grid()
plt.show()
```

```python
t = np.arange(0, 10, 0.2)
plt.plot(t, t**2, 'r--', label = 't^2')
plt.plot(t, t**2.2, 'b^', label = 't^2.2')
plt.plot(t, t**2.5, 'gs', label = 't^2.5')
plt.grid()
plt.legend()
plt.xlabel("Exponent values")
plt.ylabel("t values")
plt.title("My plot")
plt.show()
```
