```python
import pandas as pd
pakistan_weather = pd.DataFrame({
    'city':['Karachi', 'Islamabad', 'Lahore', 'Peshawar'],
    'temperture':[45, 35, 40, 37],
    'humidity':[77, 73, 72, 64]
})
pakistan_weather
```

```python
import pandas as pd
india_weather = pd.DataFrame({
    'city':['Mumbai', 'Chinai', 'Dehli', 'Gujarat'],
    'temperture':[43, 36, 30, 44],
    'humidity':[80, 60, 70, 66]
})
india_weather
```

```python
weather = pd.concat([pakistan_weather, india_weather])
weather
```

```python
weather = pd.concat([pakistan_weather, india_weather], ignore_index = True)
weather
```

```python
weather = pd.concat([pakistan_weather, india_weather], axis = 1)
weather
```

```python
import pandas as pd
pakistan_temperature = pd.DataFrame({
    'city':['Karachi', 'Islamabad', 'Lahore', 'Peshawar', 'Multan'],
    'temperture':[45, 35, 40, 37, 48]})

pakistan_temperature
```

```python
import pandas as pd
pakistan_humidity = pd.DataFrame({
    'city':['Karachi', 'Islamabad', 'Lahore', 'Peshawar'],    'humidity':[77, 73, 72, 64]
})
pakistan_humidity
```

```python
pd.merge(pakistan_temperature, pakistan_humidity, on = 'city')
```

```python
pd.merge(pakistan_temperature, pakistan_humidity, how = 'outer')
```
